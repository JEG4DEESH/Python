```python
class PatientReport:
    def __init__(self):
        self.name = None
        self.age = None
        self.sex = None
        self.occupation = None
        self.hist_injury = None
        self.pain = None
        self.hand = None
        self.ext_pain = None
        self.swelling = None
        self.swelling_lvl = None
        self.pain_fac = None
        self.mrng_cyc = None
        self.mrng_cyc_dur = None
        self.blood_test = None
        self.c_reactive_value = None
        self.ccp_antibody = None
        self.rheumatic_fac = None
        self.d_pain = None
        self.d_hand = None

    def reports(self):
        self.name = str(input("enter your name: "))
        self.age = int(input("enter your age: "))
        self.sex = str(input('Male or Female or Rather not say: '))
        self.occupation = str(input("enter your occupation: "))
        self.hist_injury = str(input("did you have any previous history of injury: "))
        self.pain = str(input('what kind of pain do you have (eg. hand pain, knee pain,...) : '))
        self.hand = str(input('which hand is painful (Left/Right): '))
        self.ext_pain = str(input('where exactly is pain (eg. wrist joints, mcp joints,...): '))
        self.swelling = str(input('did you have any swelling over the joints (yes / no): '))
        if self.swelling == 'yes':
          self.swelling_lvl = str(input('level of the swelling(Low/Medium/High): '))
        self.pain_fac = str(input('Aggravating factors of pain: '))
        self.mrng_cyc = str(input('do you have any early morning stiffness & pain (yes / no): '))
        if self.mrng_cyc == 'yes':
            self.mrng_cyc_dur = str(input('what was the duration of that: '))
        self.blood_test = str(input('did you take any blood test recently (yes / no): '))
        if self.blood_test == 'yes':
            self.c_reactive_value = int(input('what was the c-reactive protein value Mg/L: '))
            self.ccp_antibody = str(input('what about Anti CCP Antibody (positive or negative): '))
            self.rheumatic_fac = str(input('what about Rheumatoid factor (positive or negative): '))

    def conditions(self):
        if self.pain == 'hand pain':
            self.d_pain = 'hand pain'
        elif self.pain == 'knee pain':
            self.d_pain = 'knee pain'
        else:
            self.d_pain = 'other pain'

        if self.hand == 'left hand' or self.hand == 'left':
            self.d_hand = 'left lateral'
        elif self.hand == 'right hand' or self.hand == 'right':
            self.d_hand = 'right lateral'
        elif self.hand == 'both hand' or self.hand == 'both':
            self.d_hand = 'bi lateral'

    def result_report(self):
        print("\nPatient Report:")
        print('patient name               :', self.name)
        print('Age                        :', self.age)
        print('sex                        :', self.sex)
        print('occupation                 :', self.occupation)
        print('history of injury          :', self.hist_injury)
        print('pain                       :', self.d_pain)
        print('painful hand               :', self.d_hand)
        print('Exact pain spots           :', self.ext_pain)
        print('swelling issues            :', self.swelling)
        print('swelling Level             :', self.swelling_lvl)
        print('when the pain is worse     :', self.pain_fac)
        print('morning cycle              :', self.mrng_cyc)
        print('morning cycle duration     :', self.mrng_cyc_dur)
        print('recently blood test taken  :', self.blood_test)
        print('CRP value                  :', self.c_reactive_value)
        print('Anti CCP Antibody          :', self.ccp_antibody)
        print('rheumatic factor           :', self.rheumatic_fac)

    def final_result(self):
      print("\nResult:")
      if self.age<16:
         print("The patient's age suggests the possibility of juvenile idiopathic arthritis.")

      elif self.swelling.lower()=='yes' and self.mrng_cyc.lower()=='yes' and self.ccp_antibody=='positive' and self.rheumatic_fac=='positive':
        print("The patient's morning stiffness & pain suggests the possibility of rheumatoid arthritis.")

      elif self.c_reactive_value <= 10 and self.rheumatic_fac=='negative':
        print('The factor suggest a lower likelihood of active inflammation or autoimmune activity.')

      elif self.c_reactive_value >= 10 and self.rheumatic_fac=='positive':
        print('The factor suggest  possibly indicative of conditions like rheumatoid arthritis.')



patient = PatientReport()
patient.reports()
patient.conditions()
patient.result_report()
patient.final_result()
```
